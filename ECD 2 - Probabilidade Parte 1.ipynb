### Probabilidade

- A probabilidade é uma medida da incerteza associada a um evento ou conjunto de eventos.
- Serve como base para muitas áreas da estatística e é usada para modelar situações em que o resultado é incerto, mesmo se o experimento pudesse ser repetido sob condições idênticas.




#### Elementos de Probabilidade

**Conceitos Básicos**

- **Espaço amostral ($\Omega$)**:

É o conjunto de todos os possíveis resultados de um experimento ou fenômeno aleatório. $\Omega = \{\omega_1,\ldots,\omega_k, \ldots\}$.

Exemplo: Lançamento de uma moeda. $\Omega = \{\text{cara}, \text{coroa}\}$



- **Evento ($A,\ B, C, \ldots$)**:

É um subconjunto do espaço amostral.

Exemplo: $A =\{\text{cara}\}$, $B = \{\text{cara}, \text{coroa}\}$, $C = \{\text{coroa}\}$

Observação: $\Omega = \{\text{cara}, \text{coroa}\}$ é chamado **evento certo** e $\emptyset$ é chamado o **evento impossível**.





- **Probabilidade de um evento ($P(A)$)**:

Probabilidade de ocorrência do evento $A\subset\Omega$.




### Exemplo: Lançamento de um dado de 6 Faces (D6)

Considere o lançamento de um dado de 6 faces, em que o interesse está no número obtido na face superior do dado.

#### Espaço Amostral  $\Omega$

O espaço amostral $\Omega$ para o experimento é dado por:

$\Omega$ = \{1, 2, 3, 4, 5, 6\}


#### Conjuntos de Eventos

- **Evento A**: "Obter um número maior que 3 e menor que 5". $A = \{4\}$

- **Evento B**: "Obter um número par". $B = \{2, 4, 6\}$

- **Evento C**: "Obter um número menor ou igual a 6". $C = \{1, 2, 3, 4, 5, 6\}$

- **Evento D**: "Obter um número maior que 6". $D = \emptyset$




As definições de probabilidade podem ser compreendidas de diferentes maneiras. Duas das abordagens mais conhecidas são a **frequentista** e a **axiomática**, que interpretam o conceito de probabilidade de formas distintas. A seguir, detalhamos cada uma:

**Definição Frequentista**

Na interpretação frequentista, a probabilidade de um evento é a **frequência relativa** de sua ocorrência em um experimento repetido sob condições idênticas. Em termos formais, a probabilidade de um evento $A$ é definida como o limite da frequência relativa de A quando o número de experimentos $n$ tende ao infinito:


$P(A) = \displaystyle \lim_{n \to \infty} \frac{\text{Número de vezes que } A \text{ ocorre}}{n}$



Por exemplo, a probabilidade de obter "cara" ao lançar uma moeda é a proporção de vezes que "cara" aparece em um grande número de lançamentos.

#### Definição Axiomática (Kolmogorov)

A abordagem axiomática, desenvolvida por Andrey Kolmogorov em 1933, define a probabilidade de maneira abstrata, por meio de axiomas matemáticos, e não com base em experimentos ou frequências.

**Axiomas de Kolmogorov**

(i) A probabilidade de qualquer evento $A$ é sempre maior ou igual a zero.
$P(A) \geq 0$.
    
(ii)  A probabilidade do evento certo ($\Omega$) é 1.
$P(\Omega) = 1$.
    
(iii) Se $A$ e $B$ são eventos mutuamente exclusivos (ou seja, não podem ocorrer ao mesmo tempo), então a probabilidade de $A$ ou $B$ ocorrer é a soma das probabilidades de $A$ e $B$:
$P(A \cup B) = P(A) + P(B)$.




## **Operações com Eventos:**

   - **União de eventos ($A \cup B$):** Elementos de $A$, $B$, ou ambos.
   - **Intersecção de eventos ($A \cap B$):** Elementos comuns a $A$ e $B$.
   - **Diferença de eventos ($A - B$):** Elementos de $A$ que não estão em $B$.
   - **Evento complementar ($A^c$ ou $\overline{A}$):** Elementos de $\Omega$ que não estão em $A$.
   - **Leis de DeMorgan:**
     - $(A \cup B)^c = A^c \cap B^c$.
     - $(A \cap B)^c = A^c \cup B^c$.





**Propriedades da Probabilidade:**
   - $0 \leq P(A) \leq 1$.
   - $P(A \cup B) = P(A) + P(B) - P(A \cap B)$.
   
   
   
**Probabilidade Condicional:**
- $P(A|B) =\displaystyle\frac{P(A \cap B)}{P(B)}.$
   
   

**Exemplo:**

Em um escritório com computadores de diferentes tipos (laptop ou desktop) e idades (antigo ou novo), calcular a probabilidade de escolher um laptop novo.

Definimos os eventos:

$A:$ o computador é um laptop ($A^C$: o computador é um desktop)

$B:$ o computador é antigo ($B^C:$ o computador é novo)

Queremos calcular $P(A\cap B^C)$.

Imagine que foram dadas as probabilidades:

$P(A) = 0.8$

$P(B) = 0.2$

$P(A\cap B) = 0.1$

Temos que

$P(A) = P((A \cap B) \cup (A \cap B^C))$


Como $A = (A \cap B) \cup (A \cap B^C)$ e $A \cap B$ e $A \cap B^C$ são disjuntos, temos:

$P(A) = P(A \cap B) + P(A \cap B^C)$

Assim

$P(A \cap B^C) = P(A) - P(A \cap B) = 0.8 - 0.1 = 0.7$


Logo, $P(A \cap B^C) = 0.8 - 0.1 = 0.7$.

## Árvore de Probabilidades

Uma árvore de probabilidades é uma ferramenta gráfica utilizada para representar de forma visual e intuitiva os resultados possíveis de um experimento aleatório, especialmente quando há eventos dependentes ou múltiplas etapas de um processo. Ela permite organizar as probabilidades associadas a diferentes cenários e calcular probabilidades compostas de forma estruturada.


Árvore.png

**Exemplo:** Probabilidade Condicional

Considere o seguinte cenário: um aluno pode ter estudado ou não para a prova, e ele pode ter passado ou reprovado na disciplina. Vamos definir os eventos da seguinte forma:

- $A$: O aluno estudou para a prova
- $B$: O aluno foi aprovado na disciplina.

Queremos calcular $P(A \cap B)$, ou seja, a probabilidade de o aluno ter estudado e ter sido aprovado na disciplina.

**Dados fornecidos:**

- $P(A) = 0.6$: Probabilidade de o aluno ter estudado para a prova.

- $P(B) = 0.75$: Probabilidade de o aluno ser aprovado na disciplina.

- $P(B | A) = 0.9$: Probabilidade de o aluno ser aprovado na disciplina, dado que ele estudou.

- $P(A \cap B^C) = P(B^C | A) \cdot P(A) = 0.1 \times 0.6 = 0.06$: Probabilidade de o aluno ter estudado e não ter sido aprovado na disciplina.


Queremos calcular $P(A \cap B)$, ou seja, a probabilidade de o aluno ter estudado e ter passado na prova.


A probabilidade de o aluno ter estudado e ter passado na disciplina é dada por:

$P(A \cap B) = P(B | A) \cdot P(A)$

$P(A \cap B) = P(B | A) \cdot P(A) = 0.9 \times 0.6 = 0.54$.

## **Independência de Eventos**

Dois eventos $A$ e $B$ são **independentes** se a ocorrência de um não afeta a ocorrência do outro. Ou seja, a probabilidade de $A$ ocorrer dado que $B$ ocorreu é igual à probabilidade de $A$ ocorrer, independentemente de $B$. Note que $P(B)$ precisa ser não nula.

Matematicamente, a independência de eventos é definida como:

$$ P(A \cap B) = P(A)  P(B).$$


Consequentemente, se $A$ e $B$ são independentes, então

$$ P(A | B) = \frac{P(A \cap B)}{P(B)} =\frac{P(A)  P(B)}{P(B)} = P(A).$$



### Exemplo:

Se lançarmos dois dados, a probabilidade de obtermos um número 4 no primeiro dado e um número 5 no segundo dado é dada por:

$$P(\text{4 no primeiro dado e 5 no segundo dado}) $$
$$ = P(\text{4 no primeiro dado})  P(\text{5 no segundo dado}) $$
$$ = \displaystyle\frac{1}{6}\cdot\frac{1}{6}  = \frac{1}{36}$$

Como os resultados dos dois dados são independentes, podemos multiplicar as probabilidades.



## **Lei da Probabilidade Total**

A **Lei da Probabilidade Total** é uma maneira de expressar a probabilidade total de um evento $B$ como a soma das probabilidades condicionais sobre diferentes cenários ou sub-eventos que partem de uma partição do espaço amostral. Essa lei é útil quando você tem diferentes maneiras de um evento $B$ ocorrer, dependendo de diferentes condições ou eventos.

### Fórmula da Lei da Probabilidade Total:

$$P(B) = \sum_{i} P(B|A_i)  P(A_i)$$

Onde:
- $\{A_1, A_2, \dots, A_n\}$ é uma partição do espaço amostral (os eventos $A_i$ são mutuamente exclusivos e coletivamente exaustivos).
- $P(B|A_i)$ é a probabilidade de $B$ ocorrer dado que $A_i$ ocorreu (probabilidade condicional).
- $P(A_i)$ é a probabilidade de $A_i$ ocorrer (probabilidade a priori).

### Interpretação Intuitiva:

A Lei da Probabilidade Total nos diz que, para calcular a probabilidade de um evento $B$, podemos somar as probabilidades de $B$ ocorrer sob diferentes condições $A_i$, ponderadas pelas probabilidades de cada uma dessas condições $A_i$ ocorrer.

### Exemplo:
Suponha que você quer calcular a probabilidade de um paciente ser diagnosticado com uma doença, mas o diagnóstico depende de vários fatores, como o histórico do paciente (ex: paciente de risco ou não).

- $A_1$: O paciente tem histórico de risco.
- $A_2$: O paciente não tem histórico de risco.
- $P(B|A_1)$: A probabilidade do teste dar positivo dado que o paciente tem histórico de risco.
- $P(B|A_2)$: A probabilidade do teste dar positivo dado que o paciente não tem histórico de risco.

A probabilidade total $P(B)$ de um paciente testar positivo seria a soma das probabilidades de $B$ ocorrer em cada cenário, ponderadas pelas probabilidades de $A_1$ e $A_2$.

$$P(B) = P(B|A_1)  P(A_1) + P(B|A_2)  P(A_2)$$

## **Teorema de Bayes**

O **Teorema de Bayes** é uma regra fundamental da probabilidade que descreve a probabilidade de um evento, com base em informações anteriores ou conhecimento prévio. Ele permite atualizar as probabilidades à medida que novas evidências ou dados se tornam disponíveis. O teorema é amplamente usado em estatísticas, aprendizado de máquina, diagnóstico médico, entre outras áreas.

### Fórmula do Teorema de Bayes:

$$P(A|B) = \frac{P(B|A)  P(A)}{P(B)}$$

Onde:
- $P(A|B)$ é a **probabilidade condicional** de $A$ dado que $B$ ocorreu .
- $P(B|A)$ é a **probabilidade condicional** de $B$ dado que $A$ ocorreu .
- $P(A)$ é a **probabilidade marginal** de $A$.
- $P(B)$ é a **probabilidade marginal** de $B$.

### Interpretação Intuitiva:

- **$P(A)$** é o conhecimento prévio sobre $A$ antes de observar o evento $B$.
- **$P(B|A)$** é a probabilidade de observar $B$, dado que $A$ ocorreu.
- **$P(A|B)$** é a probabilidade revisada de $A$, após observar o evento $B$.

### Exemplo:
Imagine que você tem um teste médico que detecta uma doença com certa precisão.

- $P(A)$: A probabilidade de um paciente ter a doença (probabilidade a priori).
- $P(B|A)$: A probabilidade de o teste ser positivo dado que o paciente tem a doença (sensibilidade do teste).
- $P(B)$: A probabilidade de o teste ser positivo, independentemente de o paciente ter ou não a doença.

Usando o Teorema de Bayes, podemos calcular a probabilidade de que o paciente realmente tenha a doença, dado que o teste deu positivo.







## **Relação entre o Teorema de Bayes e a Lei da Probabilidade Total**

O **Teorema de Bayes** pode ser visto como uma aplicação da **Lei da Probabilidade Total**. Ao usar a probabilidade total para calcular $P(B)$, podemos reescrever a fórmula do Teorema de Bayes:

$$P(A|B) = \frac{P(B|A)  P(A)}{P(B)} = \frac{P(B|A)  P(A)}{\sum_{i} P(B|A_i)  P(A_i)}$$

A **Lei da Probabilidade Total** é utilizada para calcular $P(B)$ no denominador, considerando todas as formas possíveis de $B$ ocorrer.

# Variáveis Aleatórias

### Definição
Uma **variável aleatória** é uma função $X(\omega)$ que associa a cada elemento $\omega \in \Omega$ (espaço amostral) um número real $x = x(\omega)$.

#### Exemplos

- **Exemplo 1**: Lançamento de uma moeda duas vezes. A variável aleatória $X$ é o número de caras. $X$ pode assumir valores em \{0, 1, 2\}.



- **Exemplo 2**: Em uma linha de produção, peças são classificadas em defeituosas ou não-defeituosas. Definimos $X$ como:

$X = \begin{cases}
  1, & \text{se a peça é defeituosa} \\
  0, & \text{se a peça não é defeituosa}
  \end{cases}$



**Observação:**


**Variável Aleatória de Bernoulli** é uma variável $X$ que assume apenas dois valores (0 ou 1).





## Classificação de variáveis aleatórias

- **Discreta**: Assume valores em um conjunto finito ou infinito enumerável.

Exemplo: número de pessoas residentes em um domicílio.


- **Contínua**: Assume valores em um conjunto infinito não enumerável.

Exemplo: tempo de vida de um componente eletrônico.

## Variáveis Aleatórias Discretas


A função $p_X(x) = P(X=x)$ é chamada de **função massa de probabilidade** (f.m.p.) e deve satisfazer:


  - $p_X(x_i) \geq 0$, para todo $x_i$
  
  
  - $\sum p_X(x_i) = 1$





##### Exemplo 1: Lançamento de uma moeda honesta duas vezes e $X$ é o número de caras.

Sabemos que, ao lançar uma moeda, existem duas possibilidades em cada lançamento: cara (H, _heads_) ou coroa (T, _tails_).

Como a moeda é honesta, a probabilidade de cada evento (cara ou coroa) é $\displaystyle\frac{1}{2} $.

O espaço amostral para dois lançamentos de moeda é:

$\Omega = \{(H, H), (H, T), (T, H), (T, T)\}$

Agora, vamos definir a variável aleatória $X$, que é o número de caras. Então temos:

- $X = 0$ quando não há nenhuma cara, ou seja, o resultado é $ (T, T) $,

- $X = 1$ quando há uma cara, ou seja, os resultados são $ (H, T) $ e $ (T, H) $,

- $X = 2$ quando há duas caras, ou seja, o resultado é $ (H, H) $.

Agora vamos calcular as probabilidades de $X = 0$, $X = 1$ e $X = 2$.

1. **Para $X = 0$ (nenhuma cara):**

A única sequência que resulta em 0 caras é $ (T, T) $, com probabilidade:

$P(X = 0) = P((T, T)) = \frac{1}{4}$

2. **Para $X = 1$ (uma cara):**

As sequências que resultam em 1 cara são $(H, T)$ e $(T, H)$, com probabilidade:

$P(X = 1) = P((H, T)) + P((T, H)) = \displaystyle\frac{1}{4} + \frac{1}{4} = \frac{2}{4} = \frac{1}{2}$

3. **Para $X = 2$ (duas caras):**

A única sequência que resulta em 2 caras é $(H, H)$, com probabilidade:

$P(X = 2) = P((H, H)) = \displaystyle\frac{1}{4}$

Portanto, a função massa de probabilidade $P(X = x)$ para o número de caras $X$ é:

$\begin{array}{cccc}
\hline
x & 0 & 1 & 2 \\
\hline\\
P(X = x) & \displaystyle\frac{1}{4} & \displaystyle\frac{1}{2} & \displaystyle\frac{1}{4} \\\\
\hline
\end{array}$

##### Exemplo 3: Compra de dois computadores de um lote de 8, sendo 3 defeituosos. Calcular a função de probabilidade para o número de computadores defeituosos adquiridos.

Vamos resolver o problema introduzindo a distribuição hipergeométrica. A fórmula geral da função de probabilidade para a distribuição hipergeométrica é dada por:

$P(X = k) = \displaystyle\frac{\binom{m}{k} \binom{N-m}{n-k}}{\binom{N}{n}}$

Onde:

- $N$ é o número total de elementos (neste caso, 8 computadores),
- $m$ é o número de elementos defeituosos (neste caso, 3 defeituosos),
- $n$ é o número de elementos escolhidos (neste caso, 2 computadores),
- $k$ é o número de elementos defeituosos que queremos na amostra (a variável que queremos calcular).


Agora, vamos calcular a função de probabilidade para $k = 0, 1, 2$ (o número de defeituosos pode ser 0, 1 ou 2).

1. **Para $k = 0$ (nenhum computador defeituoso):**

$P(X = 0) = \displaystyle\frac{\binom{3}{0} \binom{5}{2}}{\binom{8}{2}} = \frac{1 \times 10}{28} = \frac{10}{28} = \frac{5}{14}$

2. **Para $k = 1$ (um computador defeituoso):**

$P(X = 1) = \displaystyle\frac{\binom{3}{1} \binom{5}{1}}{\binom{8}{2}} = \frac{3 \times 5}{28} = \frac{15}{28}$

3. **Para $k = 2$ (dois computadores defeituosos):**

$P(X = 2) = \displaystyle\frac{\binom{3}{2} \binom{5}{0}}{\binom{8}{2}} = \frac{3 \times 1}{28} = \frac{3}{28}$

Portanto, a função de probabilidade $P(X = k)$ para o número de computadores defeituosos adquiridos é:

$\begin{array}{cccc}
\hline
k & 0 & 1 & 2 \\\hline\\
P(X = k) & \displaystyle\frac{10}{28} & \displaystyle\frac{15}{28} & \displaystyle\frac{3}{28} \\\\
\hline
\end{array}$

## Variáveis Aleatórias Contínuas
Para uma variável aleatória contínua, a função $f_X(x)$ é a **função densidade de probabilidade** (f.d.p.), que deve satisfazer:

- $f_X(x) \geq 0$, $\forall x$

- $\displaystyle\int_{-\infty}^{\infty} f_X(x) dx = 1$

- $P(a < X < b) = \displaystyle\int_a^b f_X(x) dx$


### Propriedades:

- Para variáveis contínuas, $P(X = a) = 0$ para qualquer valor $a$.

- $P(X > a) = 1 - P(X \leq a)$.





##### Exemplo 1: O tempo de produção de um componente $X$ tem a f.d.p. dada por:

$f_X(x) =  \begin{cases}
  \displaystyle \frac{5-x}{4}, & \text{se } 2 < x < 4, \\
  0, & \text{caso contrário}
  \end{cases}$

a) Mostrar que $f_X(x)$ é uma f.d.p. válida.  
  
b) Calcular a probabilidade de que o tempo de produção seja menor que 3 minutos.


a) Mostrar que $f_X(x)$ é uma f.d.p. válida

- A função $f_X(x) \geq 0$ para todo $x$:

$f_X(x) = \displaystyle\frac{5 - x}{4} \quad \text{para} \quad 2 < x < 4.$

Como $5 - x$ é positivo nesse intervalo ($x < 5$) e o denominador 4 é positivo, temos que $f_X(x) \geq 0$ para $2 < x < 4$.

Fora desse intervalo, $f_X(x) = 0$, o que também satisfaz a condição.


- Devemos verificar também se $\displaystyle\int_{-\infty}^{\infty} f_X(x) \, dx = 1.$

Vamos calcular a integral de $f_X(x)$ no intervalo $(2, 4)$:

$\displaystyle\int_2^4 \displaystyle\frac{5 - x}{4} \, dx.$

Resolvendo a integral:

$\displaystyle\int_2^4 \displaystyle\frac{5 - x}{4} \, dx = \displaystyle\frac{1}{4} \int_2^4 (5 - x) \, dx=$


$\left[5x - \displaystyle\frac{x^2}{2}\right]\Big|_2^4 = \left( 5 \times 4 - \displaystyle\frac{4^2}{2} \right) - \left( 5 \times 2 - \displaystyle\frac{2^2}{2} \right).$

Simplificando:

$(20 - 8) - (10 - 2) = 12 - 8 = 4.$

Portanto:

$\displaystyle\frac{1}{4} \times 4 = 1.$


Como ambas as condições são satisfeitas, $f_X(x)$ é uma f.d.p. válida.

b) Calcular a probabilidade de que o tempo de produção seja menor que 3 minutos

A probabilidade de que o tempo de produção seja menor que 3 minutos é dada pela integral da f.d.p. no intervalo $(2, 3)$, já que $f_X(x) = 0$ fora de $(2, 4)$:

$P(X < 3) = \displaystyle\int_2^3 f_X(x) \, dx = \displaystyle\int_2^3 \displaystyle\frac{5 - x}{4} \, dx.$

Resolvendo a integral:

$P(X < 3) = \displaystyle\frac{1}{4} \displaystyle\int_2^3 (5 - x) \, dx$

Aplicando os limites de integração:

$\left[ 5x - \displaystyle\frac{x^2}{2}\right]\Big|_2^3 = \left( 5 \times 3 - \displaystyle\frac{3^2}{2} \right) - \left( 5 \times 2 - \displaystyle\frac{2^2}{2} \right).$

Simplificando:

$(15 - 4.5) - (10 - 2) = 10.5 - 8 = 2.5.$

Portanto:

$P(X < 3) = \displaystyle\frac{1}{4} \times 2.5 = 0.625.$

## Função Distribuição Acumulada (f.d.a.)

### Caso Discreto
Para uma variável aleatória discreta $X$ com f.m.p. $p_X(x)$, a f.d.a. $F_X(x)$ é dada por:

$F_X(x) = P(X \leq x) = \displaystyle\sum_{x_i \leq x} p_X(x_i)$


### Caso Contínuo
Para uma variável aleatória contínua $X$ com f.d.p. $f_X(x)$, a f.d.a. $F_X(x)$ é dada por:

$F_X(x) = P(X \leq x) = \displaystyle\int_{-\infty}^{x} f_X(t) dt$

### Propriedades da f.d.a.


- $0 \leq F_X(x) \leq 1$



- $F_X(x)$ é contínua e monótona não-decrescente



- $\displaystyle\lim_{x \to -\infty} F_X(x) = 0$ e $\displaystyle\lim_{x \to \infty} F_X(x) = 1$



##### Exemplo

A função densidade de probabilidade (f.d.p.) $f(x)$ da distribuição uniforme no intervalo $[0, 1]$ é dada por:


$f(x) = \begin{cases}
1, & \text{se } 0 \leq x \leq 1, \\
0, & \text{caso contrário.}
\end{cases}$

A função distribuição acumulada $F(x)$ é a integral da f.d.p. $f(x)$. Ou seja:

$F(x) = \int_{-\infty}^{x} f(t) \, dt$

Para a distribuição uniforme $f(x) = 1$ no intervalo $[0, 1]$, a FDA será:

$F(x) = \begin{cases}
0, & \text{se } x < 0, \\
x, & \text{se } 0 \leq x \leq 1, \\
1, & \text{se } x > 1.
\end{cases}$

Ou seja, para $0 \leq x \leq 1$, a FDA é simplesmente $F(x) = x$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo o intervalo de x
x = np.linspace(-0.5, 1.5, 500)


# Função Densidade de Probabilidade (FDP)
def fdp(x):
    return np.where((x >= 0) & (x <= 1), 1, 0)

# Função de Distribuição Acumulada (FDA)
def fda(x):
    return np.where(x < 0, 0, np.where(x <= 1, x, 1))

plt.figure(figsize=(10, 4))

# Plotando a FDP
plt.subplot(1, 2, 1)
plt.plot(x, fdp(x), label='fdp', color='red')
plt.title('Função Densidade de Probabilidade (fdp)')
plt.xlabel('x')
plt.ylabel('Densidade de Probabilidade')



# Plotando a FDA

plt.subplot(1, 2, 2)
plt.plot(x, fda(x), label='fda', color='blue')
plt.title('Função de Distribuição Acumulada (fda)')
plt.xlabel('x')
plt.ylabel('Probabilidade Acumulada')
plt.legend()


plt.legend()

# Exibindo os gráficos
plt.tight_layout()
plt.show()
```

## Esperança Matemática (Valor Esperado)
A esperança matemática é dada por:
- Caso discreto:  

$E(X) = \displaystyle\sum_{x \in R_X} x\  p_X(x)$

- Caso contínuo:  

$E(X) = \displaystyle\int_{R_X} x\  f_X(x) dx$,

em que $R_X$ é o suporte de $X$, ou seja, é o espaço de todos os possíveis valores que $X$ pode assumir.

### Propriedades da Esperança Matemática

- $E(aX) = a E(X)$

- Se $X = a$, então $E(X) = a$

- $E(X \pm a) = E(X) \pm a$




## Variância
A variância de uma variável aleatória $X$ é dada por:

$Var(X) = E\left((X - E(X))^2\right) = E(X^2) - (E(X))^2$

A raiz quadrada da variância é o desvio-padrão $\sigma = DP(X)$.

### Propriedades da Variância
- $Var(aX) = a^2 Var(X)$
- Se $X = a$, então $Var(X) = 0$
- Se $X$ e $Y$ são independentes, então $Var(aX \pm bY) = a^2 Var(X) + b^2 Var(Y)$
