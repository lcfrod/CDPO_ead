### Probabilidade

- A probabilidade  uma medida da incerteza associada a um evento ou conjunto de eventos.
- Serve como base para muitas reas da estatstica e  usada para modelar situaes em que o resultado  incerto, mesmo se o experimento pudesse ser repetido sob condies idnticas.




#### Elementos de Probabilidade

**Conceitos Bsicos**

- **Espao amostral ($\Omega$)**:

 o conjunto de todos os possveis resultados de um experimento ou fenmeno aleatrio. $\Omega = \{\omega_1,\ldots,\omega_k, \ldots\}$.

Exemplo: Lanamento de uma moeda. $\Omega = \{\text{cara}, \text{coroa}\}$



- **Evento ($A,\ B, C, \ldots$)**:

 um subconjunto do espao amostral.

Exemplo: $A =\{\text{cara}\}$, $B = \{\text{cara}, \text{coroa}\}$, $C = \{\text{coroa}\}$

Observao: $\Omega = \{\text{cara}, \text{coroa}\}$  chamado **evento certo** e $\emptyset$  chamado o **evento impossvel**.





- **Probabilidade de um evento ($P(A)$)**:

Probabilidade de ocorrncia do evento $A\subset\Omega$.




### Exemplo: Lanamento de um dado de 6 Faces (D6)

Considere o lanamento de um dado de 6 faces, em que o interesse est no nmero obtido na face superior do dado.

#### Espao Amostral  $\Omega$

O espao amostral $\Omega$ para o experimento  dado por:

$\Omega$ = \{1, 2, 3, 4, 5, 6\}


#### Conjuntos de Eventos

- **Evento A**: "Obter um nmero maior que 3 e menor que 5". $A = \{4\}$

- **Evento B**: "Obter um nmero par". $B = \{2, 4, 6\}$

- **Evento C**: "Obter um nmero menor ou igual a 6". $C = \{1, 2, 3, 4, 5, 6\}$

- **Evento D**: "Obter um nmero maior que 6". $D = \emptyset$




As definies de probabilidade podem ser compreendidas de diferentes maneiras. Duas das abordagens mais conhecidas so a **frequentista** e a **axiomtica**, que interpretam o conceito de probabilidade de formas distintas. A seguir, detalhamos cada uma:

**Definio Frequentista**

Na interpretao frequentista, a probabilidade de um evento  a **frequncia relativa** de sua ocorrncia em um experimento repetido sob condies idnticas. Em termos formais, a probabilidade de um evento $A$  definida como o limite da frequncia relativa de A quando o nmero de experimentos $n$ tende ao infinito:


$P(A) = \displaystyle \lim_{n \to \infty} \frac{\text{Nmero de vezes que } A \text{ ocorre}}{n}$



Por exemplo, a probabilidade de obter "cara" ao lanar uma moeda  a proporo de vezes que "cara" aparece em um grande nmero de lanamentos.

#### Definio Axiomtica (Kolmogorov)

A abordagem axiomtica, desenvolvida por Andrey Kolmogorov em 1933, define a probabilidade de maneira abstrata, por meio de axiomas matemticos, e no com base em experimentos ou frequncias.

**Axiomas de Kolmogorov**

(i) A probabilidade de qualquer evento $A$  sempre maior ou igual a zero.
$P(A) \geq 0$.
    
(ii)  A probabilidade do evento certo ($\Omega$)  1.
$P(\Omega) = 1$.
    
(iii) Se $A$ e $B$ so eventos mutuamente exclusivos (ou seja, no podem ocorrer ao mesmo tempo), ento a probabilidade de $A$ ou $B$ ocorrer  a soma das probabilidades de $A$ e $B$:
$P(A \cup B) = P(A) + P(B)$.




## **Operaes com Eventos:**

   - **Unio de eventos ($A \cup B$):** Elementos de $A$, $B$, ou ambos.
   - **Interseco de eventos ($A \cap B$):** Elementos comuns a $A$ e $B$.
   - **Diferena de eventos ($A - B$):** Elementos de $A$ que no esto em $B$.
   - **Evento complementar ($A^c$ ou $\overline{A}$):** Elementos de $\Omega$ que no esto em $A$.
   - **Leis de DeMorgan:**
     - $(A \cup B)^c = A^c \cap B^c$.
     - $(A \cap B)^c = A^c \cup B^c$.





**Propriedades da Probabilidade:**
   - $0 \leq P(A) \leq 1$.
   - $P(A \cup B) = P(A) + P(B) - P(A \cap B)$.
   
   
   
**Probabilidade Condicional:**
- $P(A|B) =\displaystyle\frac{P(A \cap B)}{P(B)}.$
   
   

**Exemplo:**

Em um escritrio com computadores de diferentes tipos (laptop ou desktop) e idades (antigo ou novo), calcular a probabilidade de escolher um laptop novo.

Definimos os eventos:

$A:$ o computador  um laptop ($A^C$: o computador  um desktop)

$B:$ o computador  antigo ($B^C:$ o computador  novo)

Queremos calcular $P(A\cap B^C)$.

Imagine que foram dadas as probabilidades:

$P(A) = 0.8$

$P(B) = 0.2$

$P(A\cap B) = 0.1$

Temos que

$P(A) = P((A \cap B) \cup (A \cap B^C))$


Como $A = (A \cap B) \cup (A \cap B^C)$ e $A \cap B$ e $A \cap B^C$ so disjuntos, temos:

$P(A) = P(A \cap B) + P(A \cap B^C)$

Assim

$P(A \cap B^C) = P(A) - P(A \cap B) = 0.8 - 0.1 = 0.7$


Logo, $P(A \cap B^C) = 0.8 - 0.1 = 0.7$.

**Exemplo:** Probabilidade Condicional

Considere o seguinte cenrio: um aluno pode ter estudado ou no para a prova, e ele pode ter passado ou reprovado na disciplina. Vamos definir os eventos da seguinte forma:

- $A$: O aluno estudou para a prova
- $B$: O aluno foi aprovado na disciplina.

Queremos calcular $P(A \cap B)$, ou seja, a probabilidade de o aluno ter estudado e ter sido aprovado na disciplina.

**Dados fornecidos:**

- $P(A) = 0.6$: Probabilidade de o aluno ter estudado para a prova.

- $P(B) = 0.75$: Probabilidade de o aluno ser aprovado na disciplina.

- $P(B | A) = 0.9$: Probabilidade de o aluno ser aprovado na disciplina, dado que ele estudou.

- $P(A \cap B^C) = P(B^C | A) \cdot P(A) = 0.1 \times 0.6 = 0.06$: Probabilidade de o aluno ter estudado e no ter sido aprovado na disciplina.


Queremos calcular $P(A \cap B)$, ou seja, a probabilidade de o aluno ter estudado e ter passado na prova.


A probabilidade de o aluno ter estudado e ter passado na disciplina  dada por:

$P(A \cap B) = P(B | A) \cdot P(A)$

$P(A \cap B) = P(B | A) \cdot P(A) = 0.9 \times 0.6 = 0.54$.

## **Independncia de Eventos**

Dois eventos $A$ e $B$ so **independentes** se a ocorrncia de um no afeta a ocorrncia do outro. Ou seja, a probabilidade de $A$ ocorrer dado que $B$ ocorreu  igual  probabilidade de $A$ ocorrer, independentemente de $B$. Note que $P(B)$ precisa ser no nula.

Matematicamente, a independncia de eventos  definida como:

$$ P(A \cap B) = P(A)  P(B).$$


Consequentemente, se $A$ e $B$ so independentes, ento

$$ P(A | B) = \frac{P(A \cap B)}{P(B)} =\frac{P(A)  P(B)}{P(B)} = P(A).$$



### Exemplo:

Se lanarmos dois dados, a probabilidade de obtermos um nmero 4 no primeiro dado e um nmero 5 no segundo dado  dada por:

$$P(\text{4 no primeiro dado e 5 no segundo dado}) $$
$$ = P(\text{4 no primeiro dado})  P(\text{5 no segundo dado}) $$
$$ = \displaystyle\frac{1}{6}\cdot\frac{1}{6}  = \frac{1}{36}$$

Como os resultados dos dois dados so independentes, podemos multiplicar as probabilidades.



## **Lei da Probabilidade Total**

A **Lei da Probabilidade Total**  uma maneira de expressar a probabilidade total de um evento $B$ como a soma das probabilidades condicionais sobre diferentes cenrios ou sub-eventos que partem de uma partio do espao amostral. Essa lei  til quando voc tem diferentes maneiras de um evento $B$ ocorrer, dependendo de diferentes condies ou eventos.

### Frmula da Lei da Probabilidade Total:

$$P(B) = \sum_{i} P(B|A_i)  P(A_i)$$

Onde:
- $\{A_1, A_2, \dots, A_n\}$  uma partio do espao amostral (os eventos $A_i$ so mutuamente exclusivos e coletivamente exaustivos).
- $P(B|A_i)$  a probabilidade de $B$ ocorrer dado que $A_i$ ocorreu (probabilidade condicional).
- $P(A_i)$  a probabilidade de $A_i$ ocorrer (probabilidade a priori).

### Interpretao Intuitiva:

A Lei da Probabilidade Total nos diz que, para calcular a probabilidade de um evento $B$, podemos somar as probabilidades de $B$ ocorrer sob diferentes condies $A_i$, ponderadas pelas probabilidades de cada uma dessas condies $A_i$ ocorrer.

### Exemplo:
Suponha que voc quer calcular a probabilidade de um paciente ser diagnosticado com uma doena, mas o diagnstico depende de vrios fatores, como o histrico do paciente (ex: paciente de risco ou no).

- $A_1$: O paciente tem histrico de risco.
- $A_2$: O paciente no tem histrico de risco.
- $P(B|A_1)$: A probabilidade do teste dar positivo dado que o paciente tem histrico de risco.
- $P(B|A_2)$: A probabilidade do teste dar positivo dado que o paciente no tem histrico de risco.

A probabilidade total $P(B)$ de um paciente testar positivo seria a soma das probabilidades de $B$ ocorrer em cada cenrio, ponderadas pelas probabilidades de $A_1$ e $A_2$.

$$P(B) = P(B|A_1)  P(A_1) + P(B|A_2)  P(A_2)$$

## **Teorema de Bayes**

O **Teorema de Bayes**  uma regra fundamental da probabilidade que descreve a probabilidade de um evento, com base em informaes anteriores ou conhecimento prvio. Ele permite atualizar as probabilidades  medida que novas evidncias ou dados se tornam disponveis. O teorema  amplamente usado em estatsticas, aprendizado de mquina, diagnstico mdico, entre outras reas.

### Frmula do Teorema de Bayes:

$$P(A|B) = \frac{P(B|A)  P(A)}{P(B)}$$

Onde:
- $P(A|B)$  a **probabilidade condicional** de $A$ dado que $B$ ocorreu .
- $P(B|A)$  a **probabilidade condicional** de $B$ dado que $A$ ocorreu .
- $P(A)$  a **probabilidade marginal** de $A$.
- $P(B)$  a **probabilidade marginal** de $B$.

### Interpretao Intuitiva:

- **$P(A)$**  o conhecimento prvio sobre $A$ antes de observar o evento $B$.
- **$P(B|A)$**  a probabilidade de observar $B$, dado que $A$ ocorreu.
- **$P(A|B)$**  a probabilidade revisada de $A$, aps observar o evento $B$.

### Exemplo:
Imagine que voc tem um teste mdico que detecta uma doena com certa preciso.

- $P(A)$: A probabilidade de um paciente ter a doena (probabilidade a priori).
- $P(B|A)$: A probabilidade de o teste ser positivo dado que o paciente tem a doena (sensibilidade do teste).
- $P(B)$: A probabilidade de o teste ser positivo, independentemente de o paciente ter ou no a doena.

Usando o Teorema de Bayes, podemos calcular a probabilidade de que o paciente realmente tenha a doena, dado que o teste deu positivo.







## **Relao entre o Teorema de Bayes e a Lei da Probabilidade Total**

O **Teorema de Bayes** pode ser visto como uma aplicao da **Lei da Probabilidade Total**. Ao usar a probabilidade total para calcular $P(B)$, podemos reescrever a frmula do Teorema de Bayes:

$$P(A|B) = \frac{P(B|A)  P(A)}{P(B)} = \frac{P(B|A)  P(A)}{\sum_{i} P(B|A_i)  P(A_i)}$$

A **Lei da Probabilidade Total**  utilizada para calcular $P(B)$ no denominador, considerando todas as formas possveis de $B$ ocorrer.

# Variveis Aleatrias

### Definio
Uma **varivel aleatria**  uma funo $X(\omega)$ que associa a cada elemento $\omega \in \Omega$ (espao amostral) um nmero real $x = x(\omega)$.

#### Exemplos

- **Exemplo 1**: Lanamento de uma moeda duas vezes. A varivel aleatria $X$  o nmero de caras. $X$ pode assumir valores em \{0, 1, 2\}.



- **Exemplo 2**: Em uma linha de produo, peas so classificadas em defeituosas ou no-defeituosas. Definimos $X$ como:

$X = \begin{cases}
  1, & \text{se a pea  defeituosa} \\
  0, & \text{se a pea no  defeituosa}
  \end{cases}$



**Observao:**


**Varivel Aleatria de Bernoulli**  uma varivel $X$ que assume apenas dois valores (0 ou 1).





## Classificao de variveis aleatrias

- **Discreta**: Assume valores em um conjunto finito ou infinito enumervel.

Exemplo: nmero de pessoas residentes em um domiclio.


- **Contnua**: Assume valores em um conjunto infinito no enumervel.

Exemplo: tempo de vida de um componente eletrnico.

## Variveis Aleatrias Discretas


A funo $p_X(x) = P(X=x)$  chamada de **funo massa de probabilidade** (f.m.p.) e deve satisfazer:


  - $p_X(x_i) \geq 0$, para todo $x_i$
  
  
  - $\sum p_X(x_i) = 1$





##### Exemplo 1: Lanamento de uma moeda honesta duas vezes e $X$  o nmero de caras.

Sabemos que, ao lanar uma moeda, existem duas possibilidades em cada lanamento: cara (H, _heads_) ou coroa (T, _tails_).

Como a moeda  honesta, a probabilidade de cada evento (cara ou coroa)  $\displaystyle\frac{1}{2} $.

O espao amostral para dois lanamentos de moeda :

$\Omega = \{(H, H), (H, T), (T, H), (T, T)\}$

Agora, vamos definir a varivel aleatria $X$, que  o nmero de caras. Ento temos:

- $X = 0$ quando no h nenhuma cara, ou seja, o resultado  $ (T, T) $,

- $X = 1$ quando h uma cara, ou seja, os resultados so $ (H, T) $ e $ (T, H) $,

- $X = 2$ quando h duas caras, ou seja, o resultado  $ (H, H) $.

Agora vamos calcular as probabilidades de $X = 0$, $X = 1$ e $X = 2$.

1. **Para $X = 0$ (nenhuma cara):**

A nica sequncia que resulta em 0 caras  $ (T, T) $, com probabilidade:

$P(X = 0) = P((T, T)) = \frac{1}{4}$

2. **Para $X = 1$ (uma cara):**

As sequncias que resultam em 1 cara so $(H, T)$ e $(T, H)$, com probabilidade:

$P(X = 1) = P((H, T)) + P((T, H)) = \displaystyle\frac{1}{4} + \frac{1}{4} = \frac{2}{4} = \frac{1}{2}$

3. **Para $X = 2$ (duas caras):**

A nica sequncia que resulta em 2 caras  $(H, H)$, com probabilidade:

$P(X = 2) = P((H, H)) = \displaystyle\frac{1}{4}$

Portanto, a funo massa de probabilidade $P(X = x)$ para o nmero de caras $X$ :

$\begin{array}{cccc}
\hline
x & 0 & 1 & 2 \\
\hline\\
P(X = x) & \displaystyle\frac{1}{4} & \displaystyle\frac{1}{2} & \displaystyle\frac{1}{4} \\\\
\hline
\end{array}$

##### Exemplo 3: Compra de dois computadores de um lote de 8, sendo 3 defeituosos. Calcular a funo de probabilidade para o nmero de computadores defeituosos adquiridos.

Vamos resolver o problema introduzindo a distribuio hipergeomtrica. A frmula geral da funo de probabilidade para a distribuio hipergeomtrica  dada por:

$P(X = k) = \displaystyle\frac{\binom{m}{k} \binom{N-m}{n-k}}{\binom{N}{n}}$

Onde:

- $N$  o nmero total de elementos (neste caso, 8 computadores),
- $m$  o nmero de elementos defeituosos (neste caso, 3 defeituosos),
- $n$  o nmero de elementos escolhidos (neste caso, 2 computadores),
- $k$  o nmero de elementos defeituosos que queremos na amostra (a varivel que queremos calcular).


Agora, vamos calcular a funo de probabilidade para $k = 0, 1, 2$ (o nmero de defeituosos pode ser 0, 1 ou 2).

1. **Para $k = 0$ (nenhum computador defeituoso):**

$P(X = 0) = \displaystyle\frac{\binom{3}{0} \binom{5}{2}}{\binom{8}{2}} = \frac{1 \times 10}{28} = \frac{10}{28} = \frac{5}{14}$

2. **Para $k = 1$ (um computador defeituoso):**

$P(X = 1) = \displaystyle\frac{\binom{3}{1} \binom{5}{1}}{\binom{8}{2}} = \frac{3 \times 5}{28} = \frac{15}{28}$

3. **Para $k = 2$ (dois computadores defeituosos):**

$P(X = 2) = \displaystyle\frac{\binom{3}{2} \binom{5}{0}}{\binom{8}{2}} = \frac{3 \times 1}{28} = \frac{3}{28}$

Portanto, a funo de probabilidade $P(X = k)$ para o nmero de computadores defeituosos adquiridos :

$\begin{array}{cccc}
\hline
k & 0 & 1 & 2 \\\hline\\
P(X = k) & \displaystyle\frac{10}{28} & \displaystyle\frac{15}{28} & \displaystyle\frac{3}{28} \\\\
\hline
\end{array}$

## Variveis Aleatrias Contnuas
Para uma varivel aleatria contnua, a funo $f_X(x)$  a **funo densidade de probabilidade** (f.d.p.), que deve satisfazer:

- $f_X(x) \geq 0$, $\forall x$

- $\displaystyle\int_{-\infty}^{\infty} f_X(x) dx = 1$

- $P(a < X < b) = \displaystyle\int_a^b f_X(x) dx$


### Propriedades:

- Para variveis contnuas, $P(X = a) = 0$ para qualquer valor $a$.

- $P(X > a) = 1 - P(X \leq a)$.





##### Exemplo 1: O tempo de produo de um componente $X$ tem a f.d.p. dada por:

$f_X(x) =  \begin{cases}
  \displaystyle \frac{5-x}{4}, & \text{se } 2 < x < 4, \\
  0, & \text{caso contrrio}
  \end{cases}$

a) Mostrar que $f_X(x)$  uma f.d.p. vlida.  
  
b) Calcular a probabilidade de que o tempo de produo seja menor que 3 minutos.


a) Mostrar que $f_X(x)$  uma f.d.p. vlida

- A funo $f_X(x) \geq 0$ para todo $x$:

$f_X(x) = \displaystyle\frac{5 - x}{4} \quad \text{para} \quad 2 < x < 4.$

Como $5 - x$  positivo nesse intervalo ($x < 5$) e o denominador 4  positivo, temos que $f_X(x) \geq 0$ para $2 < x < 4$.

Fora desse intervalo, $f_X(x) = 0$, o que tambm satisfaz a condio.


- Devemos verificar tambm se $\displaystyle\int_{-\infty}^{\infty} f_X(x) \, dx = 1.$

Vamos calcular a integral de $f_X(x)$ no intervalo $(2, 4)$:

$\displaystyle\int_2^4 \displaystyle\frac{5 - x}{4} \, dx.$

Resolvendo a integral:

$\displaystyle\int_2^4 \displaystyle\frac{5 - x}{4} \, dx = \displaystyle\frac{1}{4} \int_2^4 (5 - x) \, dx=$


$\left[5x - \displaystyle\frac{x^2}{2}\right]\Big|_2^4 = \left( 5 \times 4 - \displaystyle\frac{4^2}{2} \right) - \left( 5 \times 2 - \displaystyle\frac{2^2}{2} \right).$

Simplificando:

$(20 - 8) - (10 - 2) = 12 - 8 = 4.$

Portanto:

$\displaystyle\frac{1}{4} \times 4 = 1.$


Como ambas as condies so satisfeitas, $f_X(x)$  uma f.d.p. vlida.

b) Calcular a probabilidade de que o tempo de produo seja menor que 3 minutos

A probabilidade de que o tempo de produo seja menor que 3 minutos  dada pela integral da f.d.p. no intervalo $(2, 3)$, j que $f_X(x) = 0$ fora de $(2, 4)$:

$P(X < 3) = \displaystyle\int_2^3 f_X(x) \, dx = \displaystyle\int_2^3 \displaystyle\frac{5 - x}{4} \, dx.$

Resolvendo a integral:

$P(X < 3) = \displaystyle\frac{1}{4} \displaystyle\int_2^3 (5 - x) \, dx$

Aplicando os limites de integrao:

$\left[ 5x - \displaystyle\frac{x^2}{2}\right]\Big|_2^3 = \left( 5 \times 3 - \displaystyle\frac{3^2}{2} \right) - \left( 5 \times 2 - \displaystyle\frac{2^2}{2} \right).$

Simplificando:

$(15 - 4.5) - (10 - 2) = 10.5 - 8 = 2.5.$

Portanto:

$P(X < 3) = \displaystyle\frac{1}{4} \times 2.5 = 0.625.$

## Funo Distribuio Acumulada (f.d.a.)

### Caso Discreto
Para uma varivel aleatria discreta $X$ com f.m.p. $p_X(x)$, a f.d.a. $F_X(x)$  dada por:

$F_X(x) = P(X \leq x) = \displaystyle\sum_{x_i \leq x} p_X(x_i)$


### Caso Contnuo
Para uma varivel aleatria contnua $X$ com f.d.p. $f_X(x)$, a f.d.a. $F_X(x)$  dada por:

$F_X(x) = P(X \leq x) = \displaystyle\int_{-\infty}^{x} f_X(t) dt$

### Propriedades da f.d.a.


- $0 \leq F_X(x) \leq 1$



- $F_X(x)$  contnua e montona no-decrescente



- $\displaystyle\lim_{x \to -\infty} F_X(x) = 0$ e $\displaystyle\lim_{x \to \infty} F_X(x) = 1$



##### Exemplo

A funo densidade de probabilidade (f.d.p.) $f(x)$ da distribuio uniforme no intervalo $[0, 1]$  dada por:


$f(x) = \begin{cases}
1, & \text{se } 0 \leq x \leq 1, \\
0, & \text{caso contrrio.}
\end{cases}$

A funo distribuio acumulada $F(x)$  a integral da f.d.p. $f(x)$. Ou seja:

$F(x) = \int_{-\infty}^{x} f(t) \, dt$

Para a distribuio uniforme $f(x) = 1$ no intervalo $[0, 1]$, a FDA ser:

$F(x) = \begin{cases}
0, & \text{se } x < 0, \\
x, & \text{se } 0 \leq x \leq 1, \\
1, & \text{se } x > 1.
\end{cases}$

Ou seja, para $0 \leq x \leq 1$, a FDA  simplesmente $F(x) = x$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo o intervalo de x
x = np.linspace(-0.5, 1.5, 500)


# Funo Densidade de Probabilidade (FDP)
def fdp(x):
    return np.where((x >= 0) & (x <= 1), 1, 0)

# Funo de Distribuio Acumulada (FDA)
def fda(x):
    return np.where(x < 0, 0, np.where(x <= 1, x, 1))

plt.figure(figsize=(10, 4))

# Plotando a FDP
plt.subplot(1, 2, 1)
plt.plot(x, fdp(x), label='fdp', color='red')
plt.title('Funo Densidade de Probabilidade (fdp)')
plt.xlabel('x')
plt.ylabel('Densidade de Probabilidade')



# Plotando a FDA

plt.subplot(1, 2, 2)
plt.plot(x, fda(x), label='fda', color='blue')
plt.title('Funo de Distribuio Acumulada (fda)')
plt.xlabel('x')
plt.ylabel('Probabilidade Acumulada')
plt.legend()


plt.legend()

# Exibindo os grficos
plt.tight_layout()
plt.show()
```

## Esperana Matemtica (Valor Esperado)
A esperana matemtica  dada por:
- Caso discreto:  

$E(X) = \displaystyle\sum_{x \in R_X} x\  p_X(x)$

- Caso contnuo:  

$E(X) = \displaystyle\int_{R_X} x\  f_X(x) dx$,

em que $R_X$  o suporte de $X$, ou seja,  o espao de todos os possveis valores que $X$ pode assumir.

### Propriedades da Esperana Matemtica

- $E(aX) = a E(X)$

- Se $X = a$, ento $E(X) = a$

- $E(X \pm a) = E(X) \pm a$




## Varincia
A varincia de uma varivel aleatria $X$  dada por:

$Var(X) = E\left((X - E(X))^2\right) = E(X^2) - (E(X))^2$

A raiz quadrada da varincia  o desvio-padro $\sigma = DP(X)$.

### Propriedades da Varincia
- $Var(aX) = a^2 Var(X)$
- Se $X = a$, ento $Var(X) = 0$
- Se $X$ e $Y$ so independentes, ento $Var(aX \pm bY) = a^2 Var(X) + b^2 Var(Y)$
